```python
import zmq
import json
```

```python
ctx = zmq.Context()
s = ctx.socket(zmq.REQ)
```

```python
s.connect('tcp://127.0.0.1:5555')
```

```python
s.send(json.dumps({
    "type": "suggestRoutes",
    "payload": {
        "time": 8*3600*1000,
        "startLat": 1.303242,
        "startLng": 103.893944,
        "endLat": 1.299313,
        "endLng": 103.788395
    }
}).encode('utf8'))

result = s.recv()

json.loads(result.decode('utf8'))['payload']
```

```python
s.close()
```

```python
results = json.loads(result.decode('utf8'))['payload']
results = sorted(results, key=lambda x:len(x['requests']))
results
```

```python
# plot the results in a map
import folium
m = folium.Map(location=[1.38, 103.8], zoom_start=12)

interested_results = results[-3:]

for result in interested_results:
    folium.PolyLine(
        [ [stop['lat'], stop['lng']] for stop in result['stops'] ],
        opacity=0.5,
        popup='%s people' % len(result['requests'])
    ).add_to(m)
    
# get set of stops
set_of_stops = set([ (stop['lat'], stop['lng']) for stop in result['stops'] for result in interested_results])
for stop in set_of_stops:
    folium.Marker(
        stop
    ).add_to(m)
    
m
```
